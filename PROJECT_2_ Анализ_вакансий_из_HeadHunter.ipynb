```python
# PROJECT-2. Задание 4.1
query4_1 = f'''select
              a.name area,
              count(v.id) cnt
           from public.areas a
           join public.vacancies v on a.id = v.area_id
           group by a.id
           order by cnt desc
           limit 5
        '''

df['vacancies_num_in_area'] = pd.read_sql_query(query4_1, connection)
display(df['vacancies_num_in_area'])
```

```python
# PROJECT-2. Задание 4.2
query4_2 = f'''select
              count(*) cnt
           from public.vacancies v
           where (v.salary_from is not null) or (v.salary_to is not null)
           order by 1 desc
        '''

df['salary_fill_cnt'] = pd.read_sql_query(query4_2, connection)
display(df['salary_fill_cnt'])
```

```python
# PROJECT-2. Задание 4.3
query4_2 = f'''select
              avg(v.salary_from ) lower_avg_level,
              avg(v.salary_to ) upper_avg_level
           from public.vacancies v
           where (v.salary_from is not null) or (v.salary_to is not null)
        '''

df['avg_salary'] = pd.read_sql_query(query4_2, connection)
display(df['avg_salary'])
```

```python
# PROJECT-2. Задание 4.4
query4_2 = f'''select
              v.schedule,
              v.employment,
              count(v.employment) employment_cnt
           from public.vacancies v
           group by v.schedule, v.employment
           order by employment_cnt desc
        '''

df['schedule_employment_cnt'] = pd.read_sql_query(query4_2, connection)
display(df['schedule_employment_cnt'])
```

```python
# PROJECT-2. Задание 4.5
query4_2 = f'''select
              v.experience,
              count(v.id) experience_cnt
           from public.vacancies v
           group by v.experience
           order by experience_cnt desc
        '''

df['experience_cnt'] = pd.read_sql_query(query4_2, connection)
display(df['experience_cnt'])
```
